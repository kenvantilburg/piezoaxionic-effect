```python
for i,L_1 in enumerate(tqdm(vec_L_1)):
    L_1_n = L_1 * (1 - 1j * eps_L_1)
    for j,C_1 in enumerate(vec_C_1):
        try:
            optimise = L2_find(T, l,v,a,b,h_11,c_11,beta_11,k2,L_squid, R_squid, L_i, k_i, C_1, L_1_n, k_f, eta, S_flux_squid, T_2, mu_N, n_N, xi_11, zeta_11, P_nuc, N_series,N_parallel,0, L_2)
            if optimise[1] > 0:
                L_2 = optimise[0]
                arr_omega_opt[i,j,0] = optimise[1]
                continue
            else: 
                optimise = L2_find(T, l,v,a,b,h_11,c_11,beta_11,k2,L_squid, R_squid, L_i, k_i, C_1, L_1_n, k_f, eta, S_flux_squid, T_2, mu_N, n_N, xi_11, zeta_11, P_nuc, N_series,N_parallel,1, L_2)
            if optimise[1] > 0:
                L_2 = optimise[0]
                arr_omega_opt[i,j,1] = optimise[1]
                continue
            else:
                pass
        except RuntimeWarning:
            pass
```
